```python
import pandas as pd

df = pd.read_csv('boston.csv')
print("correlation coefficient:", df.corr())
print("covariance normalized by n-1:", df.cov())     
print("covariance normalized by n:", df.cov()*4/5)  
```

```python
#Question 2

import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import fetch_california_housing
house = fetch_california_housing()

house_2D = np.array(house.data).reshape(-1,1)
custom = (0,4)
scaler= MinMaxScaler(feature_range=custom)

normalized = scaler.fit_transform(house_2D)
normalized_1D = normalized.flatten()

print("Original 1D Data:")
print(house.data)
print("Normalized 2D ")
print(normalized)
print("\nNormalized 1D Data (default range 0 to 1):")
print(normalized_1D)
```
