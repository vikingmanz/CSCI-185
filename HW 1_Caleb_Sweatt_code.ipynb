```python
#Task 8

import json
import os
import datetime

new_folder = "task8-output"
os.makedirs(new_folder, exist_ok = True)
tweets = []

with open("CrimeReport.txt", "r") as infile:
    for line in infile.readlines():
        tweet = (json.loads(line))
        tweets.append(tweet)
        
sorted_tweets = sorted(tweets, key = lambda item:
                      datetime.datetime.strptime(item['created_at'], '%a %b %d %H:%M:%S +0000 %Y'))

i = 0

while i < len(sorted_tweets) - 1:
    var1 = datetime.datetime.strptime(sorted_tweets[i]['created_at'], '%a %b %d %H:%M:%S +0000 %Y')
    var2 = datetime.datetime.strptime(sorted_tweets[i+1]['created_at'], '%a %b %d %H:%M:%S +0000 %Y')
    file_name = var1.strftime("%b")+"-"+var1.strftime("%d")+"-"+\
    var1.strftime("%Y")+"-"+var1.strftime("%H")  
    file_path = os.path.join(new_folder, file_name)
    if var1.month == var2.month and var1.day\
    == var2.day and var1.hour == var2.hour: 
        with open(file_path, "a") as outfile:
            outfile.write(json.dumps(sorted_tweets[i])+'\n')
            i += 1
        if(i == len(sorted_tweets)-1):
            with open(file_path, "a") as outfile:
                outfile.write(json.dumps(sorted_tweets[i])+'\n')
    else:    
        with open(file_path, "a") as outfile:
            outfile.write(json.dumps(sorted_tweets[i]))
            i+=1
        if(i == len(sorted_tweets)-1):
            file_name = var2.strftime("%b") +"-"+var2.strftime("%d") +"-"+\
            var2.strftime("%Y")+"-"+var2.strftime("%H")
            file_path = os.path.join(new_folder, file_name)
            with open(file_path, "a") as outfile:
                outfile.write(json.dumps(sorted_tweets[i])+'\n')
                
#see jupyter home for task 8 output folder
```
